```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
```

```python
def sigmoid(z):
    f = 1 / (1 + np.exp(-z))
    return f
```

```python
x = np.arange(-2,2,0.1)
w = 0.5
```

```python
plt.plot(x, sigmoid(0.1*x))
plt.plot(x, sigmoid(0.5*x))
plt.plot(x, sigmoid(5*x))
plt.grid(1)
```

```python
def sigmoid_prime(x):
    ds = np.exp(-x)/((1+np.exp(-x))**2)
    return ds
```

```python
x = np.arange(-5,5,0.1)
labels=[]
for w in np.arange(0, 1, 0.25):
    print w
    plt.plot(x, sigmoid(x*w))
    labels.append("sigm w="+str(w))


plt.plot(x, sigmoid_prime(x)*x)
labels.append("sigm_pr")
    
plt.grid(1)
plt.legend(labels, loc='upper left',bbox_to_anchor=(1,1))
```
